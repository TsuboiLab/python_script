```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y, y_error = np.loadtxt("./hoge.txt",  unpack=True)
```

```python
t = np.linspace(x[0],x[-1],100)
```

```python
p = np.polynomial.Chebyshev.fit(x, y, 22)
```

```python
plt.figure(figsize=(12,4))
#plt.yscale('log')
#plt.ylim(10,1000)
plt.errorbar(x, y, yerr = y_error, capsize=1, fmt='.', markersize=5, zorder=1)
plt.plot(t, p(t),  lw=3, zorder=2)
```
